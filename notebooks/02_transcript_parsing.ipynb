# 02 — Transcript Parsing (Version 0.1)

Parse the four milestone transcripts (AAPL 2024Q1/Q2, MSFT 2024Q1/Q2) into a single
clean dataframe and look at basic descriptive statistics.

Run `python -m earnings_nlp.data.download_transcripts` first.

```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path.cwd().parent / "src"))

from earnings_nlp.processing.clean_transcripts import parse_transcript_files
from earnings_nlp.utils.paths import TRANSCRIPTS_RAW

files = sorted(TRANSCRIPTS_RAW.glob("*.json"))
print(f"found {len(files)} transcript files")

df = parse_transcript_files(files)
df.head()
```

```python
print("total turns:", len(df))
print()
print("turns by call:")
print(df.groupby(["ticker", "quarter"]).size())
print()
print("turns by section:")
print(df["section"].value_counts())
print()
print("turns by role:")
print(df["role"].value_counts())
```

```python
word_counts = df["text"].str.split().str.len()
print(word_counts.describe())
```

```python
from earnings_nlp.utils.paths import DATA_PROCESSED

DATA_PROCESSED.mkdir(parents=True, exist_ok=True)
df.to_csv(DATA_PROCESSED / "parsed_transcripts.csv", index=False)
print(f"wrote {len(df)} rows to {DATA_PROCESSED / 'parsed_transcripts.csv'}")
```
