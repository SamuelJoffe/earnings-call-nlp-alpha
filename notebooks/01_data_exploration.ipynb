# 01 — Data Exploration

Inspect a raw downloaded transcript JSON payload before writing any parsing logic.
Run `python -m earnings_nlp.data.download_transcripts` first so that
`data/raw/transcripts/AAPL_2024Q1.json` (etc.) exists.

```python
import json
import sys
from pathlib import Path

sys.path.insert(0, str(Path.cwd().parent / "src"))

from earnings_nlp.utils.paths import TRANSCRIPTS_RAW

with open(TRANSCRIPTS_RAW / "AAPL_2024Q1.json") as f:
    raw = json.load(f)

print(raw.get("symbol"), raw.get("quarter"))
print("number of turns:", len(raw.get("transcript", [])))
raw["transcript"][:3]
```

```python
# Look at the distinct (speaker, title) pairs to sanity-check role classification
speakers = {(t.get("speaker"), t.get("title")) for t in raw["transcript"]}
for s in sorted(speakers, key=lambda x: str(x[0])):
    print(s)
```
